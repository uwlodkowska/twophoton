```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_1s = np.array([0.17, 0.28, 0.2, 0.29, 0.03])
data_2s = np.array([0.17, 0.27, 0.16, 0.28, 0.01])
data_3s = np.array([0.21, 0.47, 0.21, 0.31,0.1])

data_learned = data_3s-data_1s

data_1d = np.array([0.3,0.31,0.19])
data_2d = np.array([0.11,0.25,0.14])
data_3d = np.array([0.17, 0.14, 0.15])

data_dumb = data_3d-data_1d

data = [data_learned, data_dumb]
print(data)
fig = plt.figure(figsize =(5,5))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])

ax.set_xticklabels(['allocentric\nmice', 'egocentric\nmice'], fontsize=25)
# Creating plot
ax.set_ylabel("similarity index increase", fontsize=25)
bp = ax.boxplot(data)
 
# show plot
```
