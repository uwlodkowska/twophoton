```python
from skimage import io
from os import path, listdir
import numpy as np
import tifffile
import csv
import numpy as np
import pandas as pd
```

```python
centroids_path ="/media/ula/DATADRIVE1/fos_gfp_tmaze/ctx_landmark/despeckle/alignment_result/aligned_despeckle/"
```

```python
xcolname = "Intensity center X (px) (ch 0)"
ycolname = "Intensity center Y (px) (ch 0)"
zcolname = "Intensity center Z (px) (ch 0)"
```

```python
landmark_first_mice = [1, 3, 4, 6,7,9,12,15,17, 18]
ctx_first_mice = [13]
landmark_session_order = ["landmark", "ctx1", "ctx2"]
ctx_session_order = ["ctx", "landmark1", "landmark2"]

max_regions_no = 4;
xy_scale = 1.19;
z_scale = 2;
```

```python
#mouse, region, order in pair, pair being compared
filename_template = "m{}r{}_{}{}_output.txt"
```

```python
tolerance = 4
```

```python
def find_common_cells(m,r, start_img_id, suff):
    filename1 = centroids_path + filename_template.format(m, r, start_img_id, suff)
    filename2 = centroids_path + filename_template.format(m, r, start_img_id+1, suff)
    print(filename1)
    if path.exists(filename1) and path.exists(filename2):
        df1 = pd.read_csv(filename1, "\t", header=1)
        df2 = pd.read_csv(filename2, "\t", header=1)
        for df in [df1, df2]:
            df[xcolname] = df[xcolname]*xy_scale
            df[ycolname] = df[ycolname]*xy_scale
            df[zcolname] = df[zcolname]*z_scale
        counter = 0
        counter = 0
        multiple_algs = 0
        for index, row1 in df1.iterrows():
            rep_counter = 0
            for index, row2 in df2.iterrows():
                dist = np.linalg.norm(row1[xcolname:zcolname]-row2[xcolname:zcolname])
                if dist < tolerance:
                    rep_counter += 1
                    if rep_counter == 1:
                        counter+=1
                    elif rep_counter == 2:
                        multiple_algs+=1
        print(filename1, filename2, "counter ", counter, "multiple_algs ", multiple_algs)
```

```python
session_order = landmark_session_order
for m in landmark_first_mice:
    for r in range(1, max_regions_no+1):
        for start_img_id in [1,2]:
            suff = "_" + session_order[start_img_id-1] + "_" + session_order[start_img_id]
            find_common_cells(m,r, start_img_id, suff)
```

```python
session_order = ctx_session_order
for m in ctx_first_mice:
    for r in range(1, max_regions_no+1):
        for start_img_id in [1,2]:
            suff = "_" + session_order[start_img_id-1] + "_" + session_order[start_img_id]
            find_common_cells(m,r, start_img_id, suff)
```
