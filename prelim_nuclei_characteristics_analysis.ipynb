```python
from skimage import io
```

```python
watershed = io.imread(fn_w)
orig = io.imread(orig)
```

```python
print(len(np.unique(watershed)))
```

```python
for_mask = long_z['Mean Intensity (ch 0)'].astype(int)
```

```python
for_mask
```

```python
import numpy.ma as ma

maskedWatershed = ma.MaskedArray(watershed, np.in1d(watershed, for_mask))
```

```python
watershed[maskedWatershed.mask == False] = 0
orig[maskedWatershed.mask == False] = 0
watershed[maskedWatershed.mask] = 255
```

```python
import tifffile
def prepare_for_tif_save(image):
    newimg = np.empty((image.shape[0], 1, image.shape[1], image.shape[2]))
    for idx, im in enumerate(image):
        newimg[idx] = np.array([im])
    return newimg.astype("uint8")
watershed = prepare_for_tif_save(watershed)
orig = prepare_for_tif_save(orig)
tifffile.imwrite(dirn+"masked_watershed.tif",watershed,imagej=True, metadata={'unit': 'pixels','axes': 'ZCYX'})
tifffile.imwrite(dirn+"masked_despeckle.tif",orig,imagej=True, metadata={'unit': 'pixels','axes': 'ZCYX'})
```
