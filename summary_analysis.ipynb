```python
import matplotlib.pyplot as plt
%matplotlib notebook
plt.ioff()
import csv, numpy as np
```

```python
path = "/media/ula/D/2p_fosgfp2020/summary.csv"

idx_front = [4,6,7]

b1b2 = []
a1b2 = []
ab_consec = []

with open(path, "r") as source:
    rdr = csv.reader( source )
    for r in rdr:
        if len(r) == 2:
            r.insert(0,r[0])
        for idx, r_ in enumerate(r):
            r[idx] = r_.replace("'", '')
        a1b2.append(r[0][1:-1].split(', ')[1:])
        ab_consec.append(r[1][1:-1].split(', ')[1:])
        b1b2.append(r[2][1:-1].split(', ')[1:])
```

```python
a1b2 = np.array(a1b2).astype(float)
ab_consec = np.array(ab_consec).astype(float)
b1b2 = np.array(b1b2).astype(float)
```

```python
fig1, axes = plt.subplots(4,1, figsize=(6,20))
axes[0].boxplot([ab_consec[:, 0], b1b2[:, 0]])
axes[1].boxplot([ab_consec[:, 1], b1b2[:, 1]])
axes[2].boxplot([a1b2[:, 0], b1b2[:, 0]])
axes[3].boxplot([a1b2[:, 1], b1b2[:, 1]])
plt.show()
```

```python
fig1, axes = plt.subplots(4,1, figsize=(6,20))
axes[0].boxplot([ab_consec[idx_front][:, 0], b1b2[:, 0]])
axes[1].boxplot([ab_consec[idx_front][:, 1], b1b2[:, 1]])
axes[2].boxplot([a1b2[idx_front][:, 0], b1b2[:, 0]])
axes[3].boxplot([a1b2[idx_front][:, 1], b1b2[:, 1]])
plt.show()
```

```python
diff = []

for a,b in zip(b1b2[idx_front][:,1], ab_consec[idx_front][:,1]):
    print(a,b)
    diff.append(a-b)
    
fig1, axes = plt.subplots()
axes.boxplot(diff)
plt.show()
```
